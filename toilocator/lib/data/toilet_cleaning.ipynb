```python
import pandas as pd
```

```python
# Import toilet data that is scraped from toilet.org.sg
toilet = pd.read_csv("toilet_scraped_final.csv")
toilet.head()
```

```python
import requests

API_KEY = ''

def getGeoCoord(address):
    """Uses Google Geocoding API to convert address of each toilet to geographic coordinates

    Parameters
    ----------
    address : str
        The address of a toilet

    Returns
    -------
    tuple
        A tuple with latitude and longitude
    """
    params = {
        'key': API_KEY,
        'address': address.replace(' ', '+')
    }

    base_url = 'https://maps.googleapis.com/maps/api/geocode/json?'
    response = requests.get(base_url, params=params)
    data = response.json()
    if data['status'] == 'OK':
        result = data['results'][0]
        location = result['geometry']['location']
        return location['lat'], location['lng']
    else:
        return

toilet['coords'] = toilet['address'].apply(getGeoCoord)
toilet.head()
```

```python
test = toilet[toilet['coords'].isna()][["address", "coords"]]
test
```

```python
toilet = toilet.dropna(subset = ['coords'])
toilet
```

```python
toilet.info()
# 7 toilets no image link
```

```python
def count_star(html):
    """Counts the number of stars in the html string

    Parameters
    ----------
    html : str
        The html string that is scraped from 

    Returns
    -------
    tuple
        A tuple with latitude and longitude
    """
    return html.count("fa fa-star")

toilet['award_int'] = toilet['award'].apply(lambda x: count_star(x))
toilet.head()

```

```python
# drop not useful columns
toilet1 = toilet.drop(columns = ['web-scraper-order', 'web-scraper-start-url', 'image_link', 'award'])
toilet1.head()
```
